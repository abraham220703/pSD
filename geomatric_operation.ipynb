```python
# prompt: please code for image rotation using open cv and plot  image result and original

import urllib.request
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Download the image
urllib.request.urlretrieve(
    "https://tse2.mm.bing.net/th?id=OIP.Pwjf0iY22yTAAXtU5jLnMAHaEo&pid=Api&P=0&h=180",
    "gfg.png"
)

# Load the image using OpenCV
img = cv2.imread("gfg.png")

# Rotate the image by 90 degrees clockwise
rows, cols = img.shape[:2]
M = cv2.getRotationMatrix2D((cols/2, rows/2), 90, 1)
rotated_img = cv2.warpAffine(img, M, (cols, rows))

# Convert images from BGR (OpenCV's default) to RGB for matplotlib
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
rotated_img_rgb = cv2.cvtColor(rotated_img, cv2.COLOR_BGR2RGB)

# Display the original and rotated images using matplotlib
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(img_rgb)
plt.title("Original Image")

plt.subplot(1, 2, 2)
plt.imshow(rotated_img_rgb)
plt.title("Rotated Image")

plt.show()
```

```python
# prompt: please make code for image Translation  using open cv and plot image result  and original

import urllib.request
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Download the image
urllib.request.urlretrieve(
    "https://tse2.mm.bing.net/th?id=OIP.Pwjf0iY22yTAAXtU5jLnMAHaEo&pid=Api&P=0&h=180",
    "gfg.png"
)

# Load the image using OpenCV
img = cv2.imread("gfg.png")

# Define the translation matrix
# tx and ty are the translation amounts in the x and y directions, respectively
tx = 50
ty = 100
M = np.float32([[1, 0, tx], [0, 1, ty]])

# Apply the translation to the image
translated_img = cv2.warpAffine(img, M, (img.shape[1], img.shape[0]))

# Convert images from BGR (OpenCV's default) to RGB for matplotlib
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
translated_img_rgb = cv2.cvtColor(translated_img, cv2.COLOR_BGR2RGB)

# Display the original and translated images using matplotlib
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(img_rgb)
plt.title("Original Image")

plt.subplot(1, 2, 2)
plt.imshow(translated_img_rgb)
plt.title("Translated Image")

plt.show()
```

```python
# prompt: please make code for Image Normalization using open cv and plot image result and original

import urllib.request
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Download the image
urllib.request.urlretrieve(
    "https://tse2.mm.bing.net/th?id=OIP.Pwjf0iY22yTAAXtU5jLnMAHaEo&pid=Api&P=0&h=180",
    "gfg.png"
)

# Load the image using OpenCV
img = cv2.imread("gfg.png")

# Normalize the image
normalized_img = cv2.normalize(img, None, 0, 255, cv2.NORM_MINMAX)

# Convert images from BGR (OpenCV's default) to RGB for matplotlib
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
normalized_img_rgb = cv2.cvtColor(normalized_img, cv2.COLOR_BGR2RGB)

# Display the original and normalized images using matplotlib
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(img_rgb)
plt.title("Original Image")

plt.subplot(1, 2, 2)
plt.imshow(normalized_img_rgb)
plt.title("Normalized Image")

plt.show()
```

```python
# prompt: please make code for reflection using open cv and plot image result and original

import urllib.request
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Download the image
urllib.request.urlretrieve(
    "https://tse2.mm.bing.net/th?id=OIP.Pwjf0iY22yTAAXtU5jLnMAHaEo&pid=Api&P=0&h=180",
    "gfg.png"
)

# Load the image using OpenCV
img = cv2.imread("gfg.png")

# Reflect the image horizontally
reflected_img = cv2.flip(img, 1)  # 1 for horizontal flip

# Convert images from BGR (OpenCV's default) to RGB for matplotlib
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
reflected_img_rgb = cv2.cvtColor(reflected_img, cv2.COLOR_BGR2RGB)

# Display the original and reflected images using matplotlib
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(img_rgb)
plt.title("Original Image")

plt.subplot(1, 2, 2)
plt.imshow(reflected_img_rgb)
plt.title("Reflected Image")

plt.show()
```
